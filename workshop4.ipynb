Exercise 2

```python
import matplotlib.pyplot as plt

years = [2017, 2018, 2019, 2020, 2021, 2022]
bears =   [8, 54, 93, 116, 137, 184]
dolphins = [150, 77, 32, 11, 6, 1]
whales =  [80, 54, 100, 76, 93, 72]
lable = ["bears","dolphins","whales"]
for i in range(len(years)):
  x=[]
  x.append(bears[i])
  x.append(dolphins[i])
  x.append(whales[i])
  plt.title(f"Population distribution in {years[i]}")
  plt.pie(x,labels=lable)
  plt.show()
```
