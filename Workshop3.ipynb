Example 7

```python
import matplotlib.pyplot as plt
import math
x = []
y = []
start = -4 * math.pi
end = 4 * math.pi
step = 0.1
i = start
while i <= end:
  x.append(i)
  y.append(math.sin(i))
  i += step
plt.plot(x, y, color="red", label="sin function")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("y = sin(x) ,-4π<= x <= 4π")
plt.grid(True)
plt.legend()
plt.show()
```

Example 8

```python
import matplotlib.pyplot as plt
import math
x=[]
y=[]
r=4
center=[2,2]
theta_start=0.0
theta_end=2*math.pi
step_size=0.125
steps=math.ceil((theta_end-theta_start)/step_size)
theta=theta_start
for i in range(0,steps):
  x.append(r*math.cos(theta)+center[0])
  y.append(r*math.sin(theta)+center[1])
  theta+=step_size;
plt.figure()
plt.plot(center[0],center[1],'.')
plt.plot(x,y,'-')
plt.grid()
plt.xlabel("X")
plt.ylabel("Y")
ax=plt.gca()
ax.set_aspect('equal',adjustable='box')
plt.show()
```

Example 10

```python
import matplotlib.pyplot as plt
import math
q=0
p=0
a=0.5
step_size=0.1
y_start = p-5
y_end = p+5
x_values=[]
y_values=[]
y=y_start
while y<= y_end:
  x=4*a*(y-p)**2+p
  x_values.append(x)
  y_values.append(y)
  y+=step_size
plt.figure()
plt.title(f"Parabola with vertex at ({p},{q}) and a={a}")
plt.plot(x_values,y_values,'-')
plt.grid()
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
import matplotlib.pyplot as plt
import math
x=[]
y=[]
x1=[]
y1=[]
r=4
a=3
b=5
center=[2,2]
theta_start=0.0
theta_end=2*math.pi
theta_end1=3*math.pi
step_size=0.125
theta=theta_start
while (theta<=theta_end):
  x.append(r*math.cos(theta)+center[0])
  y.append(r*math.sin(theta)+center[1])
  theta+=step_size;
theta=theta_start
while (theta<=theta_end1):
  x1.append(a*math.cos(theta)+center[0])
  y1.append(b*math.sin(theta)+center[1])
  theta+=step_size
plt.figure()
plt.plot(center[0],center[1],'.')
plt.plot(x,y,'-')
plt.plot(x1,y1,'-')
plt.grid()
plt.xlabel("X")
plt.ylabel("Y")
ax=plt.gca()
ax.set_aspect('equal',adjustable='box')
plt.show()
```
