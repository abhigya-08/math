```python
def checkPrime(x):
  count = 0;
  for i in range(1,x+1):
    if(x%i==0):
      count += 1
  if(count == 2):
    print("It is prime")
  else:
    print("It is not prime")

num=int(input("enter a number "))
checkPrime(num)
```

Exercise 7

```python
def conv(n):
  binary = ""
  while n > 0:
    r = n%2
    binary = binary + str(r)
    n = n//2
  print("Binary = ", binary)

num = int(input("Enter a number "))
conv(num)
```

```python
def prodDigits():
  n=int(input("enter a number"))
  prod = 1
  while n>0:
    r = n%10
    prod = prod * r
    n = n//10
  return prod

print(prodDigits())
```
