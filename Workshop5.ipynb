```python
import numpy as np
import matplotlib.pyplot as plt
""" Problem 2: Create 2-dimensional position vectors and plot them. """
origin = np.array([0,0])
V = np.array([1,1])
W = np.array([-2.0,-1])
fig, ax = plt.subplots()
# Add vectors V and W to the plot
ax.quiver(origin[0], origin[1], V[0], V[1], angles='xy', scale_units='xy', scale=1, color='r')
ax.quiver(origin[0], origin[1], W[0], W[1], angles='xy', scale_units='xy', scale=1, color='b')
ax.text(origin[0]-0.1, origin[1]+0.1, "O")
ax.text(V[0], V[1], "v")
ax.text(W[0]-0.125, W[1]-0.1, "w")
ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
origin = np.array([0,0,0])
U= np.array([1, 0.25 ,0])
V = np.array([0, 1, 0.25])
W = np.array([0.25, 0, 1])
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.quiver(origin[0],origin[1],origin[2],U[0],U[1],U[2],color='red')
ax.quiver(origin[0],origin[1],origin[2],V[0],V[1],V[2],color='blue')
ax.quiver(origin[0],origin[1],origin[2],W[0],W[1],W[2],color='yellow')
ax.text(origin[0]-0.1,origin[1],origin[2]-0.2,"O")
ax.text(U[0],U[1],U[2],"U")
ax.text(V[0],V[1],V[2],"v")
ax.text(W[0]-0.2,W[1],W[2],"W")
ax.set_xlim([-1,1])
ax.set_ylim([-1,1])
ax.set_zlim([-1,1])
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.grid()
plt.show()
```
